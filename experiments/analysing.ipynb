Interaction features from Table 2 of [Theory-Based User Search Behaviour Modelling and Understanding through Search Log Analysis](https://dl.acm.org/doi/pdf/10.1145/3698204.3716462)

```python
# average pages viewed per user session

query_wise_stats = []
for (uid, sessionid, query), group in logs.loc[logs["type"]=="searchResultGenerated"].groupby(["uid", "sessionID", "query"]):
    query_wise_stats.append([uid, sessionid, query, len(group["page"].unique())])

query_wise_stats_df = pd.DataFrame(query_wise_stats, columns=["uid", "sessionid", "query", "num_pages_visited"])
query_wise_stats_df.groupby(["uid","sessionid"])["num_pages_visited"].mean()
```

```python
# average click position per user session
logs.loc[logs["type"]=="clickedResult"].groupby(["uid", "sessionID"])["rank"].mean()
```

```python
# average queries per user session

session_wise_data = []
for (uid, sessionid), group in logs.loc[logs["type"]=="querySubmitted"].groupby(["uid", "sessionID"]):
    session_wise_data.append([uid, sessionid, len(group["query"].unique())])

session_wise_data_df = pd.DataFrame(session_wise_data, columns=["uid", "sessionid", "num_queries"])
session_wise_data_df.groupby(["uid", "sessionid"])["num_queries"].mean()
```
